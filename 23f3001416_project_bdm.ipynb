```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Excel file
file_name = "sales.xlsx"  # Replace with actual file name
df = pd.read_excel(file_name)

# Display first few rows
df.head()
df = df.iloc[:-1]  # Removes the last row
# Convert Financial Period column to string for proper plotting
df['Financial Period'] = df['Financial Period'].astype(str)

# Plot line graph
plt.figure(figsize=(12, 6))

# Plot each column
plt.plot(df['Financial Period'], df['Older B2B Model'], marker='o', label='Older B2B Model')
plt.plot(df['Financial Period'], df['Older B2C Model'], marker='s', label='Older B2C Model')
plt.plot(df['Financial Period'], df['New B2B Model'], marker='^', label='New B2B Model')
plt.plot(df['Financial Period'], df['New B2C Model'], marker='d', label='New B2C Model')

# Labels & Formatting
plt.xticks(rotation=45)
plt.xlabel("Financial Period")
plt.ylabel("Sales Count")
plt.title("Sales Trend Over Time")
plt.legend()
plt.grid(True)
plt.show()


```
